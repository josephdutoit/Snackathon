```python
class BevDataset(torch.utils.data.Dataset):
    def __init__(self, transform=None):
        self.transform = transform
        self.data_dir = '/content/hackathon_data'
        self.data_labels = os.path.join(self.data_dir, "datasets/train.txt")

        raw_txt = np.loadtxt(self.data_labels, dtype=str)
        self.df = pd.DataFrame(raw_txt, columns=['image', 'label'])


    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_path = os.path.join(self.data_dir, self.df.iloc[idx, 0])
        image = Image.open(img_path)
        label = self.df.iloc[idx, 1]
        # if self.transform:
        #     image = self.transform(image)
        return image, label
```

```python
bev_dataset = BevDataset()
dataloader = DataLoader(bev_dataset, batch_size=1, shuffle=True)


```
